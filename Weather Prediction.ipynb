First, raw data is collected from the National Weather Service at https://www.weather.gov/wrh/Climate?wfo=mfl. This dataset contains the daily averages from 2020.

```python
import pandas as pd
miami_data = pd.read_excel(r'Miami Weather.xlsx')
pan_city_data = pd.read_excel(r'Panama City Weather.xlsx')
miami_data
```

Here are the calculations for the monthly averages.

```python
months_list = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
def avg_by_month(df, month):
    total = 0
    for i in df[month]:
        # This loops through the dataframe column and adds each value
        # to the total in order to calculate the average.
        if i != '-':
            total += (i)
    return(total/len(df[month]))

miami_avgs = [avg_by_month(miami_data, i) for i in months_list]
pan_city_avgs = [avg_by_month(pan_city_data, i) for i in months_list]
print(miami_avgs)
print(pan_city_avgs)
```

Using the data from the National Weather Service, I can create a scatter plot displaying the average daily temperatures through the past year.

```python
import numpy as np
import matplotlib.pyplot as plt

y_miami = []
y_pan_city = []

def y_values(df, month, y):
    for i in df[month]:
        if i != '-': y.append(i)
        
for i in months_list:
    y_values(miami_data,i,y_miami)
    y_values(pan_city_data,i,y_pan_city)
    
x_miami = range(len(y_miami))
x_pan_city = range(len(y_pan_city))
    
plt.scatter(x_miami, y_miami, c='blue')
plt.scatter(x_pan_city, y_pan_city, c='red')
plt.legend(['Miami','Panama City'])
plt.xlabel('Month')
plt.ylabel('Temperature(F)')
plt.title("Average Daily Temperature 2020")
x_ticks = [30.5,61,91.5,122,152.5,183,203.5,234,264.5,295,325.5,356]
plt.xticks(x_ticks,labels=months_list)
plt.show
```

The following methods, calculate the standard deviation and mean for bothdatasets.

```python
import math 

def mean(x):
    return sum(x) / len(x)
def de_mean(x):
    x_bar = mean(x)
    return [x_i - x_bar for x_i in x]
def variance(x):
    n = len(x)
    deviations = de_mean(x)
    squares=[]
    for i in deviations:
        i2 = i**2
        squares.append(i2)
    return sum(squares) / (n - 1)
def standard_deviation(x):
    return math.sqrt(variance(x))

print(standard_deviation(y_miami),mean(y_miami))
print(standard_deviation(y_pan_city),mean(y_pan_city))
```

Seaborn is used to peform a linear regression analysis to predict the weather in Miami on Christmas Day. The line shows the general trend of the temperature over the months of November and December according to the 2020 weather data. The light blue area surrounding the line shows the confidence interval. 

We can predict that on Christmas Day 2021, the average temperature in Miami between 67 and 72 degrees.

```python
import seaborn as sns

sns.regplot(x = [x_miami[-60:]], y = [y_miami[-60:]])
plt.title("Miami Regression Analysis")
plt.xlabel('Day of the Year')
plt.ylabel('Temperature(F)')
```

As for Christmas Day 2021 in Panama City, we can predict the average temperature to be between 50 and 55 degrees.

```python
sns.regplot(x = [x_pan_city[-60:]], y = [y_pan_city[-60:]])
plt.title("Panama City Regression Analysis")
plt.xlabel('Day of the Year')
plt.ylabel('Temperature(F)')
```
